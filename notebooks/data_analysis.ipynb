```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../output/speeches-1.csv", delimiter="|", lineterminator="\n")
df.head()
```

```python
speakers = [re.sub(r" \([a-z]*\)", '', str(speaker)) for speaker in df.speaker.values]
print(f"Number of speakers: {len(set(speakers))}")
n_unique_last_names = [speaker.split(" ")[len(speaker.split(" "))-1] for speaker in speakers if len(speaker.split(" ")) > 1]
print(f"Number of unique last names: {len(set(n_unique_last_names))}")
```

# Different forms for names

```python
regex_exp = "(([P|p]uheenjohta[a-z|å|ä|ö]*) [A-Z|Å|Ä|Ö][a-z|å|ä|ö]*)"\
            "|([M|m]iniste[a-z|å|ä|ö]* [A-Z|Å|Ä|Ö][a-z|å|ä|ö]*)"\
            "|([P|p]ääminister[a-z|å|ä|ö]* [A-Z|Å|Ä|Ö][a-z|å|ä|ö]*)"\
            "|([E|e]dusta[a-z|å|ä|ö]* [A-Z|Å|Ä|Ö][a-z|å|ä|ö]*)"
for speech in df.speech:
    found = re.findall(regex_exp, speech)
    if found:
        matches = [match for match in found[0] if len(match) > 0]
        for match in matches:
            print(match)
```

```python
regex_exp = "(([P|p]uheenjohta[a-z|å|ä|ö]*)+ Sipil[a-z|å|ä|ö]*)"\
            "|(([M|m]iniste[a-z|å|ä|ö]*)+ Sipil[a-z|å|ä|ö]*)"\
            "|(([P|p]ääminister[a-z|å|ä|ö]*)+ Sipil[a-z|å|ä|ö]*)"\
            "|(([E|e]dusta[a-z|å|ä|ö]*)+ Sipil[a-z|å|ä|ö]*)"
all_matches = []
for speech in df.speech:
    found = re.findall(regex_exp, speech)
    if found:
        matches = [match for match in found[0] if len(match) > 0]
        all_matches.append(matches[0])
print(set(all_matches))
```
